```python
Fund = driver.find_element_by_xpath('//*[@id="menu-item-51510"]/a')
Fund.click()
```

```python
#Scraping the funding details after clicking the fund. The below url is after clicking the fund button.
url = "https://trak.in/india-startup-funding-investment-2015/"
page = requests.get(url)
page.status_code #permission for proceeding
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
more_urls = [url]

for h3_tag in soup.find_all(name="h3"):
    more_urls.append(h3_tag.find(name='a').get('href'))
```

```python
new_row_list = []
column_name = ['Sr. No.', 'Date (dd/mm/yyyy)', 'Startup Name', 'Industry/ Vertical', 'Sub-Vertical', 'City / Location', 'Investors’ Name', 'Investment Type', 'Amount (in USD)']
more_urls = set(more_urls)

urls_count = 1
for url in more_urls:
    page = requests.get(url)
    page.status_code
    soup = BeautifulSoup(page.content, 'html.parser')
    
    class_list = []
    for element in soup.find_all(class_=True):
        class_list.extend(element["class"])
    class_list = [cls for cls in class_list if 'tablepress-id-' in cls] 
    
    if len(class_list) < 1:
        skip_first_row = True
        class_list.append(None)
        for class_ in class_list:
            tbl=soup.find(name='table')

            n_rows = 0
            for tr in tbl.find_all('tr'):
                if skip_first_row == True:
                    skip_first_row = False
                    continue
                new_row = {}
                for col_id, td in enumerate(tr.find_all('td')):
                    if col_id < len(column_name):
                        new_row[column_name[col_id]] = td.text
                if not new_row == {}:
                    n_rows += 1
                    new_row_list.append(new_row)
    else:
        for class_ in class_list:
            tbl=soup.find(name='table', class_=class_)

            n_rows = 0
            for tr in tbl.find_all('tr'):
                new_row = {}
                for col_id, td in enumerate(tr.find_all('td')):
                    if col_id < len(column_name):
                        new_row[column_name[col_id]] = td.text
                if not new_row == {}:
                    n_rows += 1
                    new_row_list.append(new_row)

Funding_Deals = pd.DataFrame(new_row_list, columns=column_name)
```

7.Write a program to scrap all the available details of top 10 gaming laptops from digit.in.

```python
#Opening the chromedriver
driver = webdriver.Chrome("chromedriver.exe")
```

```python
# Opening the website digit.in
driver.get('https://www.digit.in/')
# Opening the top ten laptops
topten = driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[1]/div[4]/div[1]/div[3]/div[2]/a/p")
topten.click()
```

```python
#making a list of all the specifications
Product_Name = []
OS = []
Display = []
Processor = []
Memory = []
Weight = []
Dimension = []
Graphics_Processor = []
Price = []
URL=[]

# Scraping all the product names
name = driver.find_elements_by_xpath("//div[@class='right-container']/div/a/h3")
for i in name:
    if i.text is None :
        Product_Name.append("--") 
    else:
        Product_Name.append(i.text)
        
# Scraping the OS of all the products        
os=driver.find_elements_by_xpath("//div[@class='product-detail']/div/ul/li[1]/div/div")
for i in os:
    if i.text is None :
        OS.append("--") 
    else:
        OS.append(i.text)

# Scraping the display of all the products
display=driver.find_elements_by_xpath("//div[@class='product-detail']/div/ul/li[2]/div/div")
for i in display:
    if i.text is None :
        Display.append("--") 
    else:
        Display.append(i.text)

# Scraping the processor of all the products
processor=driver.find_elements_by_xpath("//div[@class='product-detail']/div/ul/li[3]/div/div")
for i in processor:
    if i.text is None :
        Processor.append("--") 
    else:
        Processor.append(i.text)

# Scraping the Memory of all the products
memory=driver.find_elements_by_xpath("//div[@class='product-detail']/div/ul/li[4]/div/div")
for i in memory:
    if i.text is None :
        Memory.append("--") 
    else:
        Memory.append(i.text)

# Scraping the weight of all the products
weights=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[1]")
Weight_spec=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[3]")

for i in range(len(weights)):
    if weights[i].text=='Weight':
        Weight.append(Weight_spec[i].text)
        

# Scraping the dimensions of all the products
dims=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[1]")
dims_spec=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[3]")

for i in range(len(dims)):
    if dims[i].text=='Dimension':
        Dimension.append(dims_spec[i].text)

# Scraping the Graphics processor of all the products
GPU=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[1]")
GPU_spec=driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr/td[3]")
for i in range(len(GPU)):
    if GPU[i].text=='Graphics Processor':
        Graphics_Processor.append(GPU_spec[i].text)
        
# Scraping the price of all the products
pri=driver.find_elements_by_xpath('//*[@id="buyLink"]/td[3]')
for i in pri:
    Price.append(i.text)

# Scraping the URLs of all the products
urls=driver.find_elements_by_xpath("//div[@class='full-specs']/span")
for i in urls:
    if i.get_attribute('data-href'):
        URL.append(i.get_attribute('data-href'))
```

```python
#Putting all the scraped data into a dataframe 
Specifications = pd.DataFrame({})
Specifications['Product_Name']=Product_Name
Specifications['OS']=OS
Specifications['Display']=Display
Specifications['Processor']=Processor
Specifications['Memory']=Memory
Specifications['Weight']=Weight
Specifications['Dimension']=Dimension
Specifications['Graphics Processor']=Graphics_Processor
Specifications['Price']=Price[:10]
Specifications['URL']=URL
```
